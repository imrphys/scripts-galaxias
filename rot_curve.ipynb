```python
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
#LEO EL FICHERO Y LO PARTO EN COLUMNAS
infile = "freqdbh_cinco.dat" #PARA LAS DISTINTAS CURVAS HAY DISTINTOS FICHEROS
data = np.genfromtxt(infile, comments='#')[:,0:]
```

```python
#ASOCIO VARIABLES A COLUMNAS DEL FICHERO
radius = data[:, 0]*4.5 #PASADO A kpc
omegah = data[:, 1]
nuh = data[:, 2]
sigmad = data[:, 3]
vct = data[:, 4]
vb = data[:, 5]
nub = data[:, 6]
psitot = data[:, 7]
psi = data[:, 8]
```

```python
#CALCULO COSAS
vctot = vct*220
vcb = vb*220
vhb = np.sqrt((vcb**2) + (((omegah*radius/4.5)*220)**2))
disk = np.sqrt((vctot*vctot)-(vhb*vhb))
```

```python
#PLOTEO LAS CURVAS DE ROTACION
#PARA PLOTEAR LAS DIFERENTES CURVAS HAY QUE CAMBIAR EL FICHERO DE ENTRADA
#Y DESCOMENTAR LA LINEA DE TITULO CORRESPONDIENTE

fig = plt.figure(figsize=(10, 6), dpi=200) # Create a figure of size 8x6 inches, 80 dots per inch
ax = fig.add_subplot(1,1,1)

#ax.set_title('Curva rotacin galaxia B')
#ax.set_title('Curva rotacin galaxia B modificando \u03C8$_0$ y \u03C8$_{cut}$')
#ax.set_title('Curva rotacin galaxia B modificando (r$_c$/r$_k$)$_2$')
#ax.set_title('Curva rotacin galaxia B modificando \u03C3$_b$')
ax.set_title('Curva rotacin galaxia B modificando v$_0$')


ax.set_xlabel('Radius (kpc)')
ax.set_ylabel('Vc (km/s)')
ax.set_xlim(0,20)
ax.plot(radius, vctot, color = "blue", linewidth=1.0, linestyle="-",label="Total")
ax.plot(radius, vcb, color = "green", linewidth=1.0, linestyle=":",label="Bulbo")
ax.plot(radius, omegah*(radius/4.5)*220, color = "red", linewidth=1.0, linestyle="--",label="Halo")
ax.plot(radius, disk, color = "black", linewidth=1.0, linestyle="-.",label="Disco")
ax.legend(loc='lower right')
plt.savefig("borrar.eps")
plt.show()
```
